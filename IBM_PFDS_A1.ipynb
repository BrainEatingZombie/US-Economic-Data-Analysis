```python
import pandas as pd
```

```python
from bokeh.plotting import figure, output_file, show,output_notebook
output_notebook()
```

```python
def make_dashboard(x, gdp_change, unemployment, title, file_name):
    output_file(file_name)
    p = figure(title=title, x_axis_label='year', y_axis_label='%')
    p.line(x.squeeze(), gdp_change.squeeze(), color="firebrick", line_width=4, legend="% GDP change")
    p.line(x.squeeze(), unemployment.squeeze(), line_width=4, legend="% unemployed")
    show(p)
```

Create a dataframe that contains the GDP data and display using the method head()

```python
csv_path = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/projects/coursera_project/clean_gdp.csv'
gdp_df = pd.read_csv(csv_path)
gdp_df.head()
```

Create a dataframe that contains the unemployment data. Display the first five rows of the dataframe using the method head()

```python
csv_path2 = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/projects/coursera_project/clean_unemployment.csv'
unmp_df = pd.read_csv(csv_path2)
unmp_df.head()
```

Display a dataframe where unemployment was greater than 8.5% .

```python
prcnt_df = unmp_df[unmp_df['unemployment'] > 8.5]
prcnt_df.head()
```

Use the function make_dashboard to make a dashboard

```python
x = gdp_df[['date']]
gdp_change = gdp_df[['change-current']]
unemployment = unmp_df[['unemployment']]
title = "US Economic Data Dashboard"
file_name = "index.html"
make_dashboard(x = x, gdp_change = gdp_change, unemployment = unemployment, title = title, file_name = file_name)
```
